```python
import cv2
import numpy as np
```

```python
def nothing(x):
    pass
```

```python
drawing = False # true if mouse is pressed
mode = True # if True, draw rectangle. Press 'm' to toggle to curve
ix,iy = -1,-1
ox,oy = -1,-1
#img = cv2.imread('333.jpg')

# mouse callback function
def draw_circle(event,x,y,flags,param):
    global ix,iy,ox,oy,drawing,img,mode
    #img = cv2.imread('333.jpg')
    
    if event == cv2.EVENT_LBUTTONDOWN:
        #img = cv2.imread('333.jpg')
        drawing = True
        ix,iy = x,y
    elif event == cv2.EVENT_MOUSEMOVE:
        if drawing == True:
            if mode == True:
                #img = np.zeros((512,512,3), np.uint8)
                #img = cv2.imread('333.jpg')
                cv2.rectangle(img,(ix,iy),(x,y),(0,0,255),0)
                ox,oy = x,y
    elif event == cv2.EVENT_LBUTTONUP:
        drawing = False
        if mode == True:
            #img = cv2.imread('333.jpg')
            cv2.rectangle(img,(ix,iy),(x,y),(0,0,255),0)
            ox,oy = x,y
```

```python
#img = np.zeros((512,512,3), np.uint8)
#img = cv2.imread('333.jpg')
cv2.namedWindow('image')
cv2.setMouseCallback('image',draw_circle)
cv2.createTrackbar('value','image',0,255,nothing)

while(1):
    img = cv2.imread('333.jpg')
    k = cv2.waitKey(1) & 0xFF
    
    r = cv2.getTrackbarPos('value','image')
    #img[:] = [r]
    
    color = img[iy:oy,ix:ox]
    color[:,:,0] = 0 
    color[:,:,1] = 0 
    
    font = cv2.FONT_HERSHEY_SIMPLEX
    m = "Mouse position (" + str(ix) + "," + str(iy)+ ") - (" + str(ox) + "," + str(oy) + ")"
    cv2.putText(img,m,(10,20), font, 0.5,(255,255,255),1,cv2.LINE_AA)
    
    cv2.imshow('image',img)
    
    if k == 27:
        break
cv2.destroyAllWindows()
```
